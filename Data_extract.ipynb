```python
import pandas as pd
import numpy as np
```

```python
nse_file = "nse.xlsx"
bse_file = "bse.csv"
```

```python
# Loading NSE Excel file
xl = pd.ExcelFile(nse_file)

# Print the sheet names
print(xl.sheet_names)

# Load a sheet into a DataFrame by name: df1
nse = xl.parse('MCAP_31032020')
nse = nse.drop(['Sr. No.', 'Market Capitalisation \n(In lakhs)'], axis=1)
def add_ticker_nse(x):
    return x + '.NS'
nse = nse.apply(add_ticker_nse)
```

```python
bse = pd.read_csv(bse_file)

# Removing Delisted or Suspended Companies
bse = bse[bse.Status != 'Delisted']
bse = bse[bse.Status != 'Suspended']
bse = bse.drop(['Security Code', 'Issuer Name', 'Status', 'Group', 'Face Value', 'ISIN No', 'Industry', 'Instrument'], axis=1)
bse = bse.rename(columns={'Security Id': 'Symbol', 'Security Name': 'Company Name'})
def add_ticker_bse(x):
    return x + '.BO'
bse = bse.apply(add_ticker_bse)
bse.reset_index()
```

```python
companies = pd.concat([bse, nse], ignore_index=True)
companies = companies.drop_duplicates(subset=['Symbol'], ignore_index=True)
companies = companies[companies['Symbol'].notna()]
companies.reset_index()
companies.to_csv('companies.csv', index=False)
```

```python
from datetime import timedelta
d = monthly_returns #data['Close']
time = d.index.to_list()
print(time)
dt = [str(t.day) + '-' + str(t.month) + '-' + str(t.year) for t in time ]
col = []
for date in dt:
    col.append(date)
```
